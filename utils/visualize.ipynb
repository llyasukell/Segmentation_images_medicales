```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

TASK_DIR = "../data/heart/Task02_Heart"
RAW_CASE = "la_003_0000.nii.gz"

raw_image_path = os.path.join(TASK_DIR, "imagesTr", RAW_CASE)


raw_volume = nib.load(raw_image_path).get_fdata()
print(f"Raw shape: {raw_volume.shape}")

def show_raw_slice(z):
    plt.figure(figsize=(6, 6))
    plt.imshow(raw_volume[:, :, z].T, cmap="gray", origin="lower")
    plt.title(f"Image brute — slice axiale {z+1} / {raw_volume.shape[2]}")
    plt.axis("off")
    plt.show()

interact(
    show_raw_slice,
    z=IntSlider(
        min=0,
        max=raw_volume.shape[2] - 1,
        step=1,
        value=raw_volume.shape[2] // 2
    )
)
```

```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

DATA_DIR = "../data/heart/Task02_Heart"
image_case = "la_003_0000.nii.gz"
label_case = "la_003.nii.gz"

image_path = os.path.join(DATA_DIR, "imagesTr_cropped", image_case)
label_path = os.path.join(DATA_DIR, "labelsTr_cropped", label_case)

print("image exists =", os.path.exists(image_path))
print("label exists =", os.path.exists(label_path))

image_vol = nib.load(image_path).get_fdata()
label_vol = nib.load(label_path).get_fdata()

print("Image shape:", image_vol.shape)
print("Label shape:", label_vol.shape)

def show_overlay(z):
    plt.figure(figsize=(6, 6))
    plt.imshow(image_vol[:, :, z].T, cmap="gray", origin="lower")
    plt.imshow(label_vol[:, :, z].T, cmap="Reds", origin="lower", alpha=0.35)
    plt.title(f"Image + label prétraités — slice {z+1} / {image_vol.shape[2]}")
    plt.axis("off")
    plt.show()

interact(
    show_overlay,
    z=IntSlider(min=0, max=image_vol.shape[2]-1, step=1, value=image_vol.shape[2]//2)
)
```

```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

DATA_DIR = "../data/heart/Task02_Heart"
image_case = "la_003_0000.nii.gz"
label_case = "la_003.nii.gz"

image_path = os.path.join(DATA_DIR, "imagesTr_cropped", image_case)
label_path = os.path.join(DATA_DIR, "labelsTr_cropped", label_case)

image_volume = nib.load(image_path).get_fdata()
label_volume = nib.load(label_path).get_fdata()

def show_image_and_label(z):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    axes[0].imshow(image_volume[:, :, z].T, cmap="gray", origin="lower")
    axes[0].set_title("Image")
    axes[0].axis("off")

    axes[1].imshow(label_volume[:, :, z].T, cmap="Reds", origin="lower")
    axes[1].set_title("Label")
    axes[1].axis("off")

    plt.suptitle(f"Slice axiale {z+1} / {image_volume.shape[2]}")
    plt.tight_layout()
    plt.show()

interact(
    show_image_and_label,
    z=IntSlider(min=0, max=image_volume.shape[2]-1, step=1, value=image_volume.shape[2]//2)
);
```
