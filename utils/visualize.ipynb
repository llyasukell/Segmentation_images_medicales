```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider, Dropdown
import os
```

```python
# PATH
DATA_DIR = r"data\Task02_Heart"
case = "la_003.nii.gz"
volume = nib.load(os.path.join(DATA_DIR, "imagesTr", case)).get_fdata()
print(f"Shape: {volume.shape}")

def show_slice(z):
    plt.figure(figsize=(6, 6))
    plt.imshow(volume[:, :, z].T, cmap="gray", origin="lower")
    plt.title(f"Slice axiale {z+1} / {volume.shape[2]}")
    plt.axis("off")
    plt.show()
    
interact(show_slice, z=IntSlider(min=0, max=volume.shape[2]-1, step=1, value=volume.shape[2]//2))
```
